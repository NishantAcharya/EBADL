```python
#Splitting into training and test sets:
train_Y = df[21][:691193]
test_Y = df[21][691193:]

df.drop(21,axis=1,inplace = True)
```

```python
val = d1_df.values

train_X = df[:691193]
test_X = df[691193:]
```

```python
#Reshaping the input layer
train_X_val = train_X.values.reshape(train_X.shape[0],1,train_X.shape[1])
test_X_val = test_X.values.reshape(test_X.shape[0],1,test_X.shape[1])
```

```python
train_Y_val = train_Y.values
test_Y_val = test_Y.values
```

```python
print(train_Y_val.shape)
print(train_X_val.shape)
print(test_X_val[:3600])
print(test_Y_val[:3600])
```

```python
# design network
model = Sequential()
model.add(LSTM(50, input_shape=(train_X_val.shape[1], train_X_val.shape[2]))) #input Layer
#model.add(Dense(100)) #Hidden Layer
#model.add(Dense(50)) #Hidden Layer
model.add(Dense(1))
model.compile(loss='mae', optimizer='adam')#output layer
# fit network
history = model.fit(train_X_val, train_Y_val, epochs=360, validation_data=(test_X_val, test_Y_val), verbose=2, shuffle=False)
```

```python
from matplotlib import pyplot
```

```python
pyplot.plot(history.history['loss'], label='train')
pyplot.plot(history.history['val_loss'], label='test')
pyplot.legend()
pyplot.show()
```
